**Distributions of feature values: Plot histograms for each numeric variable in the dataframe to review the range and density**

```python
#Plot histograms for 'num_inpatient_visit'
diabetes['num_inpatient_visit'].hist(bins=10, figsize=(8,5), range=[0, 10])
plt.xlabel('Number of Inpatient Visits')
plt.ylabel('Count')
```

This graph is skewed right. The majority of patients did not have inpatient visits. Just under 20,000 patients had around 1 inpatient visit, under 10,000 patients had between 1-2 inpatient visits and very few patients had between 4-10 inpatient visits.

```python
#Plot histograms for 'num_outpatient_visit'
diabetes['num_outpatient_visit'].hist(figsize=(8,5), range=[0, 8])
plt.xlabel('Number of Outpatient Visits')
plt.ylabel('Count')
```

This graph is skewed right. On the right side there are a roughly 7,500 patients who had more Non-lab procedures, between 5-6, than the rest of the patients. About 25,000 patients had between 2-4, but most of the patients had between 0-1 Non-lab procedures.

```python
#Plot histograms for 'number_diagnoses'
diabetes['number_diagnoses'].hist(bins=10, figsize=(8,5), range=[0,10])
plt.xlabel('Number of Diagnoses')
plt.ylabel('Count')
```

This graph is skewed left. On the left side there are a couple of hundred patients who had less number of diagnoses than the rest. Roughly half of the patients had between 4-8 diagnoses and the other half had around 9 diagnoses.

```python
#Plot histograms for 'time_in_hospital'
diabetes['time_in_hospital'].hist(bins=10, figsize=(8,5))
plt.xlabel('Time Spent in Hospital (# days)')
plt.ylabel('Count')
```

## The target variable: `readmitted`
**Reminder: 
Both â€˜>30 daysâ€™ and â€˜NOâ€™ readmission categories are grouped together and coded as Class 0 and â€˜<30 daysâ€™ is coded as Class 1**

Based on the function below defined in the Data Wrangling step, 88.65% of the population were not readmitted <30 days to the hospital after being discharged and around 11.36% of the population were readmitted <30 days. This represents an imbalanced classification problem.

```python
#re-plotting the graph of the numer of readmissions into the hospital.
#Set xticks labels to display 'No' and '<30' instead if 0 and 1
sns.set()
sns.countplot(data=diabetes, x='readmitted')
plt.xlabel('Readmission to Hospital')
plt.xticks([0, 1], ['No', '<30'])
plt.ylabel('Number of Records')
```

```python
def calc_prev(x):
    return(sum(x/len(x))*100)

calc_prev(diabetes['readmitted']==0), calc_prev(diabetes['readmitted']==1) 
```

**Let's analyse our main dependent variable, `readmitted`, based on Gender, Race, Age, Admission and Discharge details, A1C result and Glucose serum level.**

```python
#plot the 'readmitted' and the 'gender' variables
pd.crosstab(diabetes['readmitted'], diabetes['gender']).plot(kind='bar', figsize=(8,5), rot=45)
plt.xlabel('Readmitted')
plt.xticks([0, 1], ['No', '<30'])
plt.ylabel('Number of Records')
plt.title('Patient Readmission with Gender')
```

This plot shows that the number of females and males are almost equal between both classes.

```python
#plot the 'readmitted' and the 'race' variables
plt.figure(figsize = (8,5))
pd.crosstab(diabetes['readmitted'], diabetes['race']).plot(kind='bar', figsize=(8,5), rot=45)
plt.xlabel('Readmitted')
plt.xticks([0, 1], ['No', '<30'])
plt.ylabel('Number of Records')
plt.title('Patient Readmission with Race')
```

The data appears to show that a higher percentage of Caucasians were readmitted to the hospital, but this most likely a reflection of the individual distributions both 'race' and 'readmitted', shown in earlier graphs.

```python
#plot the 'readmitted' and the 'age' variables
pd.crosstab(diabetes['readmitted'], diabetes['age']).plot(kind='bar', figsize=(8,5), rot=45)
plt.xlabel('Readmitted')
plt.xticks([0, 1], ['No', '<30'])
plt.ylabel('Number of Records')
plt.title('Patient Readmission with Age')
```

It is not certain whether there is any correlation between patients being readmitted based on age. The distribution of age across both classes is basically the same as the  original distribution of age.

```python
#plot the 'readmitted' and the 'admission_source_id' variables
#â€˜>30â€™ and â€˜NOâ€™ readmission categories are grouped as Class 0 and â€˜<30â€™ as Class 1
pd.crosstab(diabetes['admission_source_id'], diabetes['readmitted']).plot(kind='bar', figsize=(8,5))
plt.xlabel('Admission Source')
plt.ylabel('Number of Records')
plt.title('Patient Readmission with Admission Source ID')
```

The number of patients being readmitted is highest for the 'Emergency Room', followed by 'Physician referral'.

```python
#plot the 'readmitted' and the 'admission_type_id' variables
#â€˜>30â€™ and â€˜NOâ€™ readmission categories are grouped as Class 0 and â€˜<30â€™ as Class 1
pd.crosstab(diabetes['admission_type_id'], diabetes['readmitted']).plot(kind='bar', figsize=(8,5))
plt.xlabel('Admission Type')
plt.ylabel('Number of Records')
plt.title('Patient Readmission with Admission Type ID')
```

The number of patients being readmitted is highest for 'Emergency' followed by 'Urgent' and then 'Elective'.

```python
#plot the 'readmitted' and the 'discharge_disposition_id' variables
#â€˜>30â€™ and â€˜NOâ€™ readmission categories are grouped as Class 0 and â€˜<30â€™ as Class 1
pd.crosstab(diabetes['discharge_disposition_id'], diabetes['readmitted']).plot(kind='bar',figsize = (8,5))
plt.xlabel('Discharge Disposition')
plt.ylabel('Number of Records')
plt.title('Patient Readmission with Discharge Disposition ID')
```

The number of patients being readmitted is highest for 'Discharge to home', followed equally by 'Discharged/transferred to SNF' and 'Discharged/transferred to home with home health service'.

```python
#plot the 'readmitted' and the 'A1Cresult' variables
#â€˜>30â€™ and â€˜NOâ€™ readmission categories are grouped as Class 0 and â€˜<30â€™ as Class 1
sns.set()
plt.figure(figsize = (8,5))
sns.countplot(x='A1Cresult', hue='readmitted', data=diabetes)
plt.xlabel('A1C')
plt.ylabel('Number of Records')
plt.title('Patient Readmission with A1C test result')
```

A1C levels are a strong indicator of Glucose in the body. The majority of the records in our dataset are marked as â€˜Noneâ€™. This is not ideal since A1C is a strong predictor of diabetes in patients. Hopefully, we'll be able to reconcile these records for our analysis.

```python
#plot the 'readmitted' and the 'max_glu_serum' variables
#â€˜>30â€™ and â€˜NOâ€™ readmission categories are grouped as Class 0 and â€˜<30â€™ as Class 1
sns.set()
plt.figure(figsize = (8,5))
sns.countplot(x='max_glu_serum', hue='readmitted', data=diabetes)
plt.xlabel('Glucose')
plt.ylabel('Number of Records')
plt.title('Patient Readmission with Glucose test result')
```
